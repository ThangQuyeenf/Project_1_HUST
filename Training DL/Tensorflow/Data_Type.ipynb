## **5.2. Linked list**

Linked List là chuỗi các phần tử mà tất cả được kết nối với nhau thông qua các liên kết(link). Mỗi phần tử ngoài việc lưu trữ giá trị của mình, nó còn lưu thêm vị trí của phần tử kế tiếp nó.

Bản chất ta có thể thấy Linked List là chuỗi được ghép bởi các Node ta đã thực hiện ở phía trên

image.png

```python
class Node:
    def __init__(self, val):
        self.Data = val
        self.next = None
number1 = Node(1)
number2 = Node(2)
number3 = Node(3)
type(number1)
```

```python
class LinkedList:
  def __init__(self):
    self.head = None
  # Hàm in tất cả các giá trị trong list
  def printVal(self):
    cur_Node =  self.head
    while cur_Node:
      print(cur_Node.Data)
      cur_Node = cur_Node.next
  def appendNode(self,node):
    if isinstance(node, Node):
      cur_Node = self.head
      # Lặp qua tất cả các Node
      # Nếu Node tiếp theo tồn tại thì di chuyển vị trí hiện tại sang Node tiếp theo
      # Nếu Node tiếp theo không hề tồn tại thì thêm Node mới
      while cur_Node:
        if cur_Node.next is not None:
          cur_Node = cur_Node.next
        else:
          cur_Node.next = node
          break
    else:
      raise ValueError("This is not  Node type.")
  def size(self):
    # Trả về độ dài của list
    count = 0
    cur_Node = self.head
    while cur_Node is not None:
      count += 1
      cur_Node = cur_Node.next
    return count
```

```python
# Khởi tạo 1 linked list
mylist = LinkedList()
# Đặt number1 làm head của list 
mylist.head = number1
# Nối Node đầu với Node thứ hai
number1.next = number2
# in giá trị của list
mylist.printVal()
```

```python
# Thêm một Node number3 vào cuối list
mylist.appendNode(number3)
# check
mylist.printVal()
```

**Bài 1: Convert Binary Number in a Linked List to Integer**

Chuyển 1 Linked List nhị phân sang số nguyên hệ 10

Ý tưởng: Duyệt qua từng Node tại mỗi nút tính giá trị hệ 10 Node đó


image.png

```python
# Hàm xây linked list tử mảng
def buidLinkedListFromArray(arr):
  linkedlist = LinkedList()
  if len(arr) == 0:
    return linkedlist
  else:
    head = Node(arr[0])
    linkedlist.head = head
    for i in range(1, len(arr)):
      head.next = Node(arr[i])
      head= head.next
  return linkedlist
```

```python
inp = [1,0,0,1,1]
binaryList = buidLinkedListFromArray(inp)
binaryList.printVal()
```

```python
# Hàm chuyển đổi binary linked list sang integer
def convertBinaryLinkedListToInteger(llist):
  total = 0
  cur_Node = llist.head
  while cur_Node:
    total = total*2 + cur_Node.Data
    cur_Node = cur_Node.next
  return total
```

```python
convertBinaryLinkedListToInteger(binaryList)
```

**Bài 2:  Middle of the Linked List**

Given a non-empty, singly linked list with head node head, return a middle node of linked list.

If there are two middle nodes, return the second middle node.

Cho 1 Danh sách liên kết không trống với head, Trả về nút giữa (tâm) của danh sách

Nếu có hai nút giữa, trả về nút giữa thứ hai






image.png

```python
arr = [1,5,6,8,7,6,9]
```

```python
middleList = buidLinkedListFromArray(arr)
```

```python
# Hàm tìm tâm của danh sách liên kết
def getMiddleOfLinkedList(llist):
  i = llist.head
  j = llist.head
  count = 0
  while i and i.next:
    i = i.next.next
    j = j.next
    count += 1
  return j, count
```

```python
# giá trị tại tâm
getMiddleOfLinkedList(middleList)[0].Data
```

```python
# Index của tâm
getMiddleOfLinkedList(middleList)[1]
```

## **6.2. Push and Pop**

image.png

**Push** là thêm một phần tử lên trên cùng của ngăn xếp

**Pop** là lấy đi phần tử trên cùng của ngăn xếp

```python
class Stack:
  def __init__(self):
    self.stack = []
  # Thêm 1 phần tử lên trên cùng của ngăn xếp 
  def Push(self, element):
    self.stack.append(element)
  # Xóa đi phần tử trên cùng của ngăn xếp
  def Pop(self):
    if len(self.stack) > 0:
      self.stack.pop()
    else:
      print("Stack is Empty")
  
  # In tất cả các phần tử của ngăn xếp
  def printStack(self):
    if len(self.stack) > 0:
      for elem in self.stack:
        print(elem,end = ' ')
    else:
      print("Stack is Empty") 
```

```python
stack2 = Stack()
stack2.printStack()
```

```python
stack2.Push(4)
stack2.printStack()
```

```python
stack2.Push(34)
stack2.Push(-1)
stack2.printStack()
```

```python
stack2.Pop()
```

```python
stack2.printStack()
```

- Duyệt qua từng phần tử của chuỗi đó 
- Nếu phần tử đó là một trong 3 ký tự mở ngoặc "(", "{", "[" thì thêm(push) vào stack
- Nếu phần tử đó là một trong 3 ký tự đóng ngoặc ")", "}", "]" thì kiểm tra phần tử cuối cùng trong stack có phải là ký tự mở ngoặc tưng ứng không:
> - nếu thỏa mãn thì xóa(pop) phần tử cuối khỏi stack
> - nếu không thỏa mãn thỉ return False
> - Chú ý khi xóa(pop) phần tử ra phải kiểm tra stack luôn có phần tử, nếu stack không có phần tử thì return False
- Chạy hết vòng lặp thì kiểm tra số lượng phần tử còn lại trong stack:
> - nếu stack vẫn còn phần tử return False
> - nếu stack trống thì return True

```python
def isValid(s: str) -> bool:
  stack = []
  mapping = {")" : "(", "]" : "[", "}" : "{"}
  for letter in s:
    if letter not in mapping:
      stack.append(letter)
    else:
      if len(stack) > 0 and stack[-1] == mapping[letter]:
        stack.pop()
      else:
        return False
  return len(stack) == 0
```

```python
isValid("()")
```

```python
isValid("()()()")
```

```python
isValid("{](")
```

## **7.2. Enqueue và Dequeue**

**Enqueue** là thêm một phần tử vào Queue, phần tử này sẽ nằm cuối cùng trong Queue

**Dequeue** là xóa đi một phần tử của Queue va phần tử này sẽ là phần tử đầu tiên của Queue
image.png

```python
class Queue:
  def __init__(self):
    self.queue = []
  # thêm một phần tử vào queue
  def enqueue(self, element):
    self.queue.append(element)
  # xóa một phần tử khỏi queue
  def dequeue(self):
    if len(self.queue) > 0:
      self.queue.pop(0)
    else:
      print("Queue is empty")
  # in các phần tử của Queue
  def printQueue(self):
    if len(self.queue) > 0:
      for elem in self.queue:
        print(elem, end= " ")
    else:
      print("Queue is empty")
```

```python
queue2 = Queue()
# Thêm 4 phần tử vào Queue
queue2.enqueue("Monday")
queue2.enqueue("Tuesday")
queue2.enqueue("Wednesday")
queue2.enqueue("Thursday")
```

```python
queue2.printQueue()
```

```python
queue2.dequeue()
queue2.printQueue()
```

```python
numbers_graph_2 = {
    "1": ["4", "5"],
    "2": ["3", "4"],
    "3": ["2", "5"],
    "4": ["1", "2", "5"],
    "5": ["1", "3", "4"]
}
```

```python
class Graph:
  def __init__(self,graph):
    if graph is None:
      graph = {}
    self.graph = graph
  
  def edges(self):
    edges = []
    nodes = self.graph.keys()
    # Lặp qua tất cả các node
    for node in nodes:
      # to_nodes là các đỉnh mà đỉnh hiện tại trỏ tới
      to_nodes = self.graph[node]
      for to_node in to_nodes:
        # Liệt kê tất cả các cạnh có thể xảy ra
        edge = set({node, to_node})
        # Thêm các cạnh duy nhất(Unique) vào list
        if edge not in edges:
          edges.append(edge)
    return edges

  def addEdge(self, edge):
    if type(edge) is set:
      nodes =self.graph.keys()
      # chỉ thêm edge không tồn tại từ trước
      if edge not in self.edges():
        (edge_start, edge_end) = tuple(edge)
        # TH1: Đỉnh đầu và dỉnh cuối đều tồn tại
        if edge_start in nodes  and edge_end in nodes:
          self.graph[edge_start].append(edge_end)
          self.graph[edge_end].append(edge_start)
        # TH2: Đỉnh đầu tồn tại và đỉnh cuối không tồn tại
        elif edge_start in nodes and edge_end not in nodes:
          self.graph[edge_start].append(edge_end)
          self.graph[edge_end] = [edge_start]
        # TH3: Đỉnh đầu không tồn tại và đỉnh cuối tồn tại
        elif edge_start not in nodes and edge_end in nodes:
          self.graph[edge_start] = [edge_end]
          self.graph[edge_end].append(edge_start)
        # TH4: Cả hai đỉnh không tồn tại
        else:
          self.graph[edge_start] = [edge_end]
          self.graph[edge_end] = [edge_start]
      else:
        print("This edge is existed")
    else:
      print("Edge type must be set")
```

```python
class Node:
  def __init__(self,value):
    self.value = value
    self.left = None
    self.right = None

  def ___repr__(self):
    return (str(self.value))

def buildTree(data):
  if len(data) == 0:
    return Node()
  i = 0
  n = len(data)
  root = Node(data[0])
  def insertNode(root, i, n):
    left_idx = 2*i + 1
    right_idx = 2*i + 2
    if left_idx < n:
      left_value = data[left_idx]
      left_node = Node(left_value)
      root.left = insertNode(left_node, left_idx, n)

    if right_idx < n:
      right_value = data[right_idx]
      right_node = Node(right_value)
      root.right = insertNode(right_node, right_idx, n)
    return root 
  tree = insertNode(root, 0, n)
  return tree
```

## **12.3 Binary Search Tree**

```python
class Node:
    def __init__(self,data):
        self.right=self.left=None
        self.data = data
class BinarySearchTree:
    def insert(self,root,data):
        if root==None:
            return Node(data)
        else:
            if data <= root.data:
                cur=self.insert(root.left,data)
                root.left=cur
            else:
                cur=self.insert(root.right,data)
                root.right=cur
        return root

    def getHeight(self,root):
        if root == None or (root.left == None and root.right == None):
            return 0
        else:
            return 1 + max(self.getHeight(root.left),self.getHeight(root.right))

    def levelOrder(self,root):
        #Write your code here
        h = self.getHeight(root) + 1
        for i in range(1, h + 1):
            self.printGivenLevel(root,i)
    def printGivenLevel(self,root,level):
        if root == None:
            return
        if level == 1:
            print(root.data, end = ' ')
        elif level > 1:
            self.printGivenLevel(root.left, level - 1)
            self.printGivenLevel(root.right, level - 1)
```
